Summary :
Exploratory Data Analysis to study various columns and their spread, to study corelation between different columns and use plots  to answer questions such as:
<br>
1. Do students in more popular majors make more money?
2. Do students that majored in subjects that were majority female make more money?
3. Is there any link between the number of full-time employees and median salary?

```python
#getting ready for plotting in pandas
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
```

```python
#count of rows of raw data

raw_data_count = recent_grads.iloc[:,0].count()
print(raw_data_count)

```

```python
#dropping rows with missing values:
recent_grads = recent_grads.dropna()

#count of rows of cleaned data

cleaned_data_count = recent_grads.iloc[:,0].count()
print(cleaned_data_count)
```

Notice the raw and clean data that one row has been removed


```python
#creating various scatter plots
recent_grads.plot(x = 'Sample_size', y ='Median', kind = 'scatter')
```

<h1> Pandas Histogram <h1>

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(0,4):
    ax = fig.add_subplot(4,1,r+1)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```

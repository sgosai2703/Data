```python
#loading json file
with open("SAFI.json") as f:
    js_f = json.load(f)
```

```python
print(type(js_f[0]))
```

```python
#printing 1st element of list
print(js_f[0])
```

```python
#loading 1st element of list to a string object with indentation of 2 for better reading and
#using sort_keys to print the keys in ascending order
print(json.dumps(js_f[0], indent =2, sort_keys = True))
```

```python
#accessing and filtering particular values in json files. If we want to access "D_curr_crop": "maize" in the first record

print(json.dumps(js_f[0]["D_plots"][0]["D_crops"][0]["D_curr_crop"], indent =2, sort_keys = True))
```

```python
#accessing the value of "D_curr_crop" in all the records

for farms in js_f:
    plot = farms['D_plots']
    for crops in plot:
        crop = crops['D_crops']
        for curr_crops in crop:
            print(curr_crops['D_curr_crop'])
```

```python
#Being able to drill down in this way is very useful in helping you get a feel for the JSON data structure. 
#D_curr_crop may not exist in any particular dictionary within D_crops
#any of the lists D_plots or D_crops could be missing or just empty lists ([])


for farms in js_f:
    if 'D_plots' in farms :
        plot = farms['D_plots']
        for crops in plot:
            if 'D_crops' in crops :
                crop = crops['D_crops']
                for curr_crops in crop:
                    if 'D_curr_crop' in curr_crops:
                        print(curr_crops['D_curr_crop'])
```

```python
#printing unique values by adding the print result to set

unique_set = set()

for farms in js_f:
    if 'D_plots' in farms :
        plot = farms['D_plots']
        for crops in plot:
            if 'D_crops' in crops :
                crop = crops['D_crops']
                for curr_crops in crop:
                    if 'D_curr_crop' in curr_crops:
                        unique_set.add(curr_crops['D_curr_crop'])

```

```python
print(unique_set)
```

```python
#We can accumulate this information as we move through the list of dictionary objects. At the top level, farm,
#there is a unique identifier A03_quest_no which we can use. for the plot and the crop within the plot we will
#create our own simple indexing system (plot_no and crop_no). 
#At the end instead of just printing the crop name, we also print the details of where this crop is being grown
```

```python
# finding the level of nesting or position of "A03_quest_no" --->farm no.
print(json.dumps(js_f[0]["A03_quest_no"], indent =2, sort_keys = True))
```

```python
#finding the number of plots
plot_no = 0
for x in js_f[0]["D_plots"]:
    plot_no +=1
print(plot_no)
```

```python
for farms in js_f:
    plot_no = 0
    id = farms['A03_quest_no']
    if 'D_plots' in farms :
        plot = farms['D_plots']
        for crops in plot:
            crop_no = 0
            plot_no += 1
            if 'D_crops' in crops :
                crop = crops['D_crops']
                for curr_crops in crop:
                    crop_no += 1
                    if 'D_curr_crop' in curr_crops:
                        print("Farm no ", id," grows ", curr_crops['D_curr_crop']," in plot", plot_no , " and it is crop number ", crop_no)
                
                
               
```

```python
#writing analysis results to csv file

import csv
filename = "SAFIcrops.csv"
fw = open(filename, "w+")
cf = csv.writer(fw, lineterminator = '\n')

#writing header
cf.writerow(["crop", "farm_no", "plot_no", "crop_no"])

for farms in js_f:
    plot_no = 0
    id = farms["A03_quest_no"]
    if "D_plots" in farms:
        plot = farms["D_plots"]
        for x in plot:
            plot_no+=1
            crop_no =0
            if "D_crops" in x:
                y = x["D_crops"]
                for z in y:
                    crop_no += 1
                    if "D_curr_crop" in z:
                        crop = z["D_curr_crop"]
                            
                        cf.writerow([crop, id, plot_no, crop_no])
                
    
```
